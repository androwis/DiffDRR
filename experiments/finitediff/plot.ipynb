```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("runs.csv")
df["grad_diff"] = df["finite_diff"] - df["auto_diff"]
df.head()
```

```python
sns.set_context("talk")
ax = sns.relplot(
    data=df,
    x="finite_diff",
    y="grad_diff",
    col="param",
    col_wrap=3,
    linewidth=0,
    marker="o",
    alpha=0.25,
    facet_kws=dict(sharex=False, sharey=False)
)
ax.set(xlabel="FiniteDiff Gradient", ylabel="Autograd - FiniteDiff")
plt.show()
```
