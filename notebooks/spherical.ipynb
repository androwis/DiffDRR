```python
import matplotlib.pyplot as plt
import torch

from src import Detector, Siddon

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
for i in range(9):
    
    detector = Detector(
        sdr    =  1,
        theta  =  (i / 8) * torch.pi,
        phi    =  (0 / 8) * torch.pi,
        gamma  =  (0 / 8) * torch.pi,
        bx     =  10,
        by     = -10,
        bz     = -40,
        device =  device,
    )

    source, rays = detector.make_xrays(5, 5, .75, .75)
    source_ = source.detach().cpu()
    rays_ = rays.permute(2, 0, 1).detach().cpu()

    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')

    ax.scatter(source_[0]        , source_[1]        , source_[2]        , label="Source")
    ax.scatter(rays_[0].flatten(), rays_[1].flatten(), rays_[2].flatten(), label="Targets")

    ax.set(xlabel="x", ylabel="y", zlabel="z")
    ax.set(xlim=[8,12], ylim=[-12,-8], zlim=[-42,-38])

    # ax.view_init(30, 90)

    plt.legend()
    plt.show()
```
